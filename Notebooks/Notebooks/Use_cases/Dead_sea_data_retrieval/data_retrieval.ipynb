```python
import ee

# Initialize the Earth Engine library
ee.Initialize()

# Define the ROI for the Dead Sea
dead_sea_roi = ee.Geometry.Rectangle([34.7, 30.9, 35.7, 31.8])

# Define the years of interest
years = [1983, 2003, 2023]

def fetch_image(year):
    """Retrieve an image for a specific year"""
    # Filter the satellite images based on the year and the region of interest
    image = (ee.ImageCollection('LANDSAT/LT05/C01/T1')
             .filterDate(f'{year}-01-01', f'{year}-12-31')
             .filterBounds(dead_sea_roi)
             .median())  # Use median to remove clouds and other artifacts

    # Return the image with visualization parameters
    return image.visualize(bands=['B3', 'B2', 'B1'], min=0, max=255)

# Fetch and export images
for year in years:
    image = fetch_image(year)
    task = ee.batch.Export.image.toDrive(
        image=image.clip(dead_sea_roi),
        description=f'dead_sea_{year}',
        folder='Dead_Sea_Images',
        scale=30,
        fileFormat='GeoTIFF'
    )
    task.start()

print("Image export tasks started. Please check your Google Drive folder 'Dead_Sea_Images' once they're completed.")
```
