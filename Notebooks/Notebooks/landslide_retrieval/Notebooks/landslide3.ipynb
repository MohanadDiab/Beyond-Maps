```python
import ee

def export_image_to_drive(lat, lon, start_date, end_date, filename, region_radius, scale):
    # Initialize the Earth Engine library
    ee.Initialize()

    # Define the region of interest
    region = ee.Geometry.Point([lon, lat])
    square = region.buffer(region_radius).bounds()

    # Fetch the Sentinel-2 image collection
    collection = (ee.ImageCollection('COPERNICUS/S2')
                  .filterDate(ee.Date(start_date), ee.Date(end_date))
                  .filterBounds(square))

    # Check if there are any images available after filtering
    count = collection.size().getInfo()
    if count == 0:
        print(f"No images available for {filename}.")
        return

    # Take the first image from the collection and clip it to the region
    image = ee.Image(collection.first()).clip(square)

    # Select the desired bands
    bands = ['B2', 'B3', 'B4', 'B5', 'B6', 'B7', 'B8', 'B8A', 'B11', 'B12']
    image = image.select(bands)

    # Set up the export parameters
    export_params = {
        'image': image,
        'description': filename,
        'scale': scale,
        'region': square.getInfo()['coordinates'],
        'folder': 'Your_Google_Drive_Folder_Name',  # Change to your desired folder
        'fileFormat': 'GeoTIFF'
    }

    # Start the export
    task = ee.batch.Export.image.toDrive(**export_params)
    task.start()

    print(f"Export started for {filename}. Check your Google Drive once the export is complete.")
```
