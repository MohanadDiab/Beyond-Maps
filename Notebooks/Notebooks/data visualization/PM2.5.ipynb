```python
import folium
from folium.plugins import HeatMap

# Define the coordinates and PM2.5 concentration values
data = {
    "Via Santa Marta, Milan, Italy": {"coords": [45.462231, 9.183522], "pm2.5": 14.18},
    "Sensor Oggiaro, Milan, Italy": {"coords": [45.485895, 9.190435], "pm2.5": 14.77},
    "Sensor Corso 22 Marzo, Milan, Italy": {"coords": [45.462082, 9.210209], "pm2.5": 10.05},
    "Sensor Piazzale Libia, Milan, Italy": {"coords": [45.454005, 9.209494], "pm2.5": 4.27}
}

# Initialize the map centered around Milan
m = folium.Map(location=[45.464664, 9.188540], zoom_start=12)

# Prepare the heatmap data, considering PM2.5 values
heatmap_data = []
for info in data.values():
    # Each location is repeated in proportion to its PM2.5 value to simulate concentration
    repeats = int(info["pm2.5"])  # This is a simple approach; adjust the scaling as needed
    heatmap_data.extend([info["coords"]] * repeats)

# Add the heatmap layer
HeatMap(heatmap_data).add_to(m)

m
```
